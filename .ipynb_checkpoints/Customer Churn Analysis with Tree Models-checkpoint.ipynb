#### Creating Cohorts based on Tenure

##### This helps answer:

"Are newer customers more likely to churn?"

"Does customer loyalty increase over time?"

```python
no_churn = df.groupby(['Churn','tenure']).count().transpose()['No']
yes_churn = df.groupby(['Churn','tenure']).count().transpose()['Yes']
```

```python
# calculating the churn rate (%) per tenure month.
churn_rate = 100 * yes_churn / (no_churn+yes_churn)
```

```python
churn_rate.transpose()['customerID']
```

##### A plot below showing churn rate per months of tenure.

```python
plt.figure(figsize=(10,4), dpi=200)
churn_rate.iloc[0].plot()
plt.ylabel('Churn Percentage')
```

#### Broader Cohort Groups
From the existing tenure column values, i created a new column called Tenure Cohort to form 4 separate categories:
   * '0-12 Months'
   * '24-48 Months'
   * '12-24 Months'
   * 'Over 48 Months'    

```python
def cohort(tenure):
    if tenure < 13:
        return '0-12 Months'
    elif tenure < 25:
        return '12-24 Months'
    elif tenure < 49:
        return '24-48 Months'
    else:
        return "Over 48 Months"
```

```python
df['Tenure Cohort'] = df['tenure'].apply(cohort)
```

```python
df.head(10)[['tenure','Tenure Cohort']]
```

# Phase 4: Predictive Modeling

#### Goal here is to explore 4 different tree based methods:
   * A Single Decision Tree
   * Random Forest
   * AdaBoost
   * Gradient Boosting

##### Single Decision Tree

```python
# Separating out the data into X features and Y label
# Creating dummy variables where necessary and noting which features are not useful and should be dropped
X = df.drop(['Churn','customerID'],axis=1)
X = pd.get_dummies(X,drop_first=True)
```

```python
y = df['Churn']
```

```python
# train test split
from sklearn.model_selection import train_test_split 
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.10,random_state=42)
```

#### For each model we :

   1. Train a the tree model.
   2. Evaluate performance metrics 

```python
# Training a single decision tree model
from sklearn.tree import DecisionTreeClassifier
dt= DecisionTreeClassifier(max_depth=6,random_state=42)
```

```python
dt.fit(X_train,y_train)
```

```python
pred= dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,ConfusionMatrixDisplay,classification_report
```

```python
print(classification_report(y_test,pred))
```

```python
ConfusionMatrixDisplay.from_estimator(dt, X_test, y_test, cmap='Blues')
```

```python
# a DataFrame (imp_feats) that shows how important each feature is to the decision tree model
imp_feats = pd.DataFrame(data=dt.feature_importances_,index=X.columns,columns=['Feature Importance']).sort_values("Feature Importance")
```

```python
plt.figure(figsize=(14,6), dpi=200)
sns.barplot(
    data=imp_feats.sort_values('Feature Importance'),
    x=imp_feats.sort_values('Feature Importance').index,
    y='Feature Importance', hue= imp_feats.sort_values('Feature Importance').index, legend=False,
    palette='Spectral'
)
plt.xticks(rotation=90)
plt.title("Feature Importance for Decision Tree");
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,8),dpi=150)
plot_tree(dt,filled=True,feature_names=X.columns);
```
